## Modelling

```python
# Membagi dataset menjadi data latih dan data uji

from sklearn.model_selection import train_test_split

X = df_application_train_cleaned
y = df_application_train_cleaned['TARGET']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### With Logistic Regression

```python
# Impor pustaka yang diperlukan
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Membuat model Logistic Regression
lr_model = LogisticRegression()

# Melatih model dengan data latih
lr_model.fit(X_train, y_train)

# Memprediksi kelas data uji
lr_y_pred = lr_model.predict(X_test)

# Mengukur akurasi model
accuracy = accuracy_score(y_test, lr_y_pred)
print(f'Akurasi: {accuracy:.2f}')

# Menampilkan metrik evaluasi lainnya
print(classification_report(y_test, lr_y_pred))

# Menampilkan matriks konfusi
lr_conf_matrix = confusion_matrix(y_test, lr_y_pred)
print('Matriks Konfusi:')
print(lr_conf_matrix)
```

### With Naive Bayes

```python
# Impor pustaka yang diperlukan
from sklearn.naive_bayes import GaussianNB

# Membuat model Naive Bayes (Gaussian Naive Bayes)
nb_model = GaussianNB()

# Melatih model dengan data latih
nb_model.fit(X_train, y_train)

# Memprediksi kelas data uji
nb_y_pred = nb_model.predict(X_test)

# Mengukur akurasi model
accuracy = accuracy_score(y_test, nb_y_pred)
print(f'Akurasi: {accuracy:.2f}')

# Menampilkan metrik evaluasi lainnya
print(classification_report(y_test, nb_y_pred))

# Menampilkan matriks konfusi
nb_conf_matrix = confusion_matrix(y_test, nb_y_pred)
print('Matriks Konfusi:')
print(nb_conf_matrix)
```
